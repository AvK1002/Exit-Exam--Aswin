```python
# 1. Calculating the mean, median, and standard deviation for area and perimeter
area_stats = {
    "Mean": data["Area"].mean(),
    "Median": data["Area"].median(),
    "Standard Deviation": data["Area"].std()
}

perimeter_stats = {
    "Mean": data["Perimeter"].mean(),
    "Median": data["Perimeter"].median(),
    "Standard Deviation": data["Perimeter"].std()
}
print("area_stats :",area_stats)
print("perimeter_stats :",perimeter_stats)
```

```python
# 2. Calculate range and IQR for "MajorAxisLength" and "MinorAxisLength"
major_axis_range = data["MajorAxisLength"].max() - data["MajorAxisLength"].min()
major_axis_iqr = np.percentile(data["MajorAxisLength"], 75) - np.percentile(data["MajorAxisLength"], 25)

minor_axis_range = data["MinorAxisLength"].max() - data["MinorAxisLength"].min()
minor_axis_iqr = np.percentile(data["MinorAxisLength"], 75) - np.percentile(data["MinorAxisLength"], 25)

axis_stats = {
    "MajorAxisLength": {"Range": major_axis_range, "IQR": major_axis_iqr},
    "MinorAxisLength": {"Range": minor_axis_range, "IQR": minor_axis_iqr}
}
print(major_axis_range)
print(major_axis_iqr)
print(minor_axis_range)
print(minor_axis_iqr)
```

```python
# 3. Count unique Class categories
unique_classes = data["Class"].nunique()
print('Their are', unique_classes, 'unique values')
```

```python
# 4. Calculate the frequency distribution of each Class
class_frequency = data["Class"].value_counts()
print('The frequency distribution is', class_frequency)
```

```python
# 5. Scatter plot and correlation between area and perimeter
correlation_area_perimeter = data["Area"].corr(data["Perimeter"])

plt.figure(figsize=(8, 6))
sns.scatterplot(x="Area", y="Perimeter", data=data)
plt.title("Scatter Plot of Area vs. Perimeter")
plt.xlabel("Area")
plt.ylabel("Perimeter")
plt.show()
```

```python
# 6. Heatmap of correlations between selected shape features
shape_features = ["Compactness", "ShapeFactor1", "AspectRation"]
correlation_matrix = data[shape_features].corr()
print("Correlation between Area and Perimeter:", correlation_area_perimeter)
```
